```python
from sklearn.impute import SimpleImputer  # help to fill missing values
from sklearn.compose import ColumnTransformer

cat_imputer = SimpleImputer(strategy='constant', fill_value='missing')
door_imputer = SimpleImputer(strategy='constant', fill_value=4)
num_imputer = SimpleImputer(strategy='median')

cat_feat = ['Make', 'Colour']
door_feat = ['Doors']
num_feat = ['Odometer (KM)']

imputer = ColumnTransformer([
('cat_imputer', cat_imputer, cat_feat),
('door_imputer', door_imputer, door_feat),
('num_imputer', num_imputer, num_feat)
])

filled_X = imputer.fit_transform(X)
filled_X
```

```python
df_fill = pd.DataFrame(filled_X, columns=['Make', 'Colour', 'Odometer (KM)', 'Doors'])
df_fill.head()
```

```python
df_fill.isna().sum()
```

```python
from sklearn.pipeline import Pipeline
```

```python
df_miss
```

```python
cat_feat = ['Make', 'Colour']
cat_transformer = Pipeline(steps=[('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
                                  ('one_hot', OneHotEncoder(handle_unknown='ignore'))])

door_feat = ['Doors']
door_transformer = Pipeline(steps=[('imputer', SimpleImputer(strategy='constant', fill_value=4))])

num_feat = ['Odometer (KM)']
num_transformer = Pipeline(steps=[('imputer', SimpleImputer(strategy='mean'))])

preprocessor = ColumnTransformer(transformers=[('cat', cat_transformer, cat_feat),
                                               ('door', door_transformer, door_feat),
                                               ('num', num_transformer, num_feat)])

model = Pipeline(steps=[('preprocessor', preprocessor),
                        ('model', RandomForestRegressor())])

X = df_miss.drop('Price', axis=1)
y = df_miss['Price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV

pipe_grid = {'preprocessor__num__imputer__strategy': ['mean', 'median'],
             'model__n_estimators': [100, 1000],
             'model__max_depth': [None, 5],
             'model__max_features': [None],
             'model__min_samples_split': [2,4]}

# model = RandomForestRegressor()

gs_model = GridSearchCV(model,
                        pipe_grid,
                        cv=5,
                        verbose=2)

gs_model.fit(X_train, y_train)
```

```python
gs_model.score(X_test, y_test)
```
