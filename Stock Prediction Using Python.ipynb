```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

```python
# Read the data from Yahoo Finance website directly
stocks = ['AAPL', 'GOOG', 'MSFT', 'AMZN']
data = yf.download(stocks, start='2021-01-01', end='2022-01-01')
data
```

```python
# Perform cleaning
df = data['Adj Close']
df
```

```python
# What was the change in stock price over time?
df.plot(figsize=(10, 6))
plt.title('Stock Price Over Time', fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.show()
```

```python
# Visualize the change in a stock’s volume being traded, over time
volume_df = data['Volume']
volume_df.plot(figsize=(10, 6))
plt.title('Stock Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.show()
```

```python
# What was the moving average of various stocks?
moving_avg = df.rolling(window=30).mean()  # rolling window of size 30 days
moving_avg.plot(figsize=(10, 6))
plt.title('Moving Average of Stocks')
plt.xlabel('Date')
plt.ylabel('Moving Average')
plt.show()
```

```python
# What was the daily return average of a stock?
daily_return = df.pct_change()   # percentage change
daily_return.plot(figsize=(10, 6))
plt.title('Daily Return of Stocks')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.show()
```

```python
# Add a new column 'Trend' whose values are based on the 'Daily Return'
df['Trend'] = np.where(daily_return['AAPL'] > 0, 'Up', 'Down')
df.head(10)
```

```python
# Visualize trend frequency through a Pie Chart
trend_counts = df['Trend'].value_counts()
trend_counts.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))
plt.title('Trend Frequency')
plt.show()
```

```python
# What was the correlation between the daily returns of different stocks?
sns.heatmap(daily_return.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Daily Returns of Different Stocks')
plt.show()
```
